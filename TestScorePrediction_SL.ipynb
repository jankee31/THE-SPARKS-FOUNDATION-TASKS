## Task 1: Prediction using Supervised ML

### GRIP: THE SPARKS FOUNDATION

*In this task, I have tried to predict the % of marks that the students are expected to score based on the number of hours they studies.*

*This is simple linear regression task as it involves only 2 variables.*

**Step 1: Import required libraries**

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

**Step 2: Read Data from the Source**

```python
url = r"https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
s_data = pd.read_csv(url)
print("Data import successful")
s_data.head(10)
```

**Step 3: Data Visualization**

```python
# Plotting the distribution of scores
s_data.plot(x='Hours', y='Scores', style='o')  
plt.grid(b=True, color='aqua', alpha=0.3, linestyle='-.'), 
plt.title('Hours X Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

*From the above graph we can svisualize that there is a linear relation between the number of hours studied and percentage of score.*

**Step 4: Data Processing**

```python
#division of data into Attributes(inputs) and labels(outputs).
X = s_data.iloc[:, :-1].values  
y = s_data.iloc[:, 1].values
```

**Step 5: Model Training**

```python
#Splitting the data into training and testing sets, and training the algorithm.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
regressor = LinearRegression()  
regressor.fit(X_train.reshape(-1,1), y_train) 
print("Training complete.")
```

**Step 6: Plotting Regression Line**

```python
line = regressor.coef_*X+regressor.intercept_
# Plotting for the test data
plt.scatter(X, y)
plt.grid(b=True, color='aqua', alpha=0.3, linestyle='-.'), 
plt.plot(X, line,color='green')
plt.title('Hours X Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

**Step 7: Making Predictions**

```python
# Testing data
print(X_test)
# Model Prediction 
y_pred = regressor.predict(X_test)
```

**Step 8: Comparision of Actual and Predicted Model result**

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}) 
df
```

```python
#Estimating training and test score
print("Training Score:",regressor.score(X_train,y_train))
print("Test Score:",regressor.score(X_test,y_test))
```

```python
# Plotting the Bar graph to visualize the difference between the actual and predicted value
df.plot(kind='bar',figsize=(5,5))
plt.grid(which='major', linewidth='0.5', color='green')
plt.grid(which='minor', linewidth='0.5', color='red')
plt.show()
```

```python
# Testing the model with provided data : 9.25 hrs/day
hours = 9.25
test = np.array([hours])
test = test.reshape(-1, 1)
own_pred = regressor.predict(test)
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

**Step 9: Model Evaluation**

The final step is to evaluate the performance of algorithm. This step is important to compare how well different algorithms perform on a particular dataset. Here different errors have been calculated to compare the model performance and predict the accuracy.

```python
from sklearn import metrics  
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_pred)) 
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R-2:', metrics.r2_score(y_test, y_pred))
```

*R-2 gives the score of model fit and in this case we have 0.9454906892105354 which is actually a great score for this model*
